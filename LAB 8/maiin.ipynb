```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
#label encoder
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()

df['Gender'] = lb.fit_transform(df['Gender'])
```

```python
X = df[['Gender','Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
#Splitting dataset
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()

model.fit(X_train,y_train)
#checking accuracy

print(model.score(X_train, y_train))
score_nb =model.score(X_test, y_test)
print(score_nb)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'sigmoid')
model.fit(X_train,y_train)
print(model.score(X_train, y_train))
score_svm_sigmoid =model.score(X_test, y_test)
print(score_svm_sigmoid)
```

```python
model = SVC(kernel = 'rbf')
model.fit(X_train,y_train)
print(model.score(X_train, y_train))
score_svm_rbf =model.score(X_test, y_test)
print(score_svm_rbf)
```

```python
model = SVC(kernel = 'poly')
model.fit(X_train,y_train)
print(model.score(X_train, y_train))
score_svm_poly =model.score(X_test, y_test)
print(score_svm_poly)
```

```python

from sklearn.tree import DecisionTreeClassifier
cls = DecisionTreeClassifier()
cls.fit(X_train, y_train)
#checking accuracy
print(cls.score(X_train, y_train))
score_dt = cls.score(X_test, y_test)
print(score_dt)
```

```python
import matplotlib.pyplot as plt
xpoints = np.array(["score_svm_sigmoid", "score_svm_rbf", "score_svm_poly", "score_dt", "score_nb"])
ypoints = np.array([score_svm_sigmoid, score_svm_rbf, score_svm_poly, score_dt, score_nb])

plt.plot(xpoints, ypoints, marker = 'o')
plt.show()
```
